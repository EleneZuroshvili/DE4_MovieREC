```python
import pandas as pd

df = pd.read_csv("movies_metadata.csv")


df
```

```python
df = df[df['original_language'] == "en"]
df = df[df["vote_count"].notna()]
df = df.sort_values("vote_count", ascending=False).iloc[:10000]
```

```python
df
```

```python
import pandas as pd
import json
import ast
import chromadb
from tqdm import tqdm
from collections import defaultdict

# Filter rows
df = df[df["overview"].notna()].copy()
print(f"{len(df)} movies with overview loaded")

# Parse genres
def parse_genres(genres_str):
    try:
        if pd.isna(genres_str):
            return []

        # Handle strings like:
        # "[{'id': 18, 'name': 'Drama'}, {'id': 80, 'name': 'Crime'}]"
        genres = ast.literal_eval(genres_str)

        if not isinstance(genres, list):
            return []

        return [g["name"] for g in genres if isinstance(g, dict) and "name" in g]
    except Exception:
        return []

tqdm.pandas(desc="Parsing genres")
df["genre_names"] = df["genres"].progress_apply(parse_genres)

# Optional quick sanity check
print(df[["genres", "genre_names"]].head())

# Prepare documents
documents = df["overview"].tolist()
ids = [f"movie_{i}" for i in range(len(df))]

metadatas = []
genre_to_ids = defaultdict(list)

# Build metadata and genre index
for i, (_, row) in enumerate(
    tqdm(df.iterrows(), total=len(df), desc="Building metadata + genre index")
):
    movie_id = ids[i]
    genres = row["genre_names"]

    metadatas.append({
        "title": row["original_title"],
        "original_language": row["original_language"],
        "genres_json": json.dumps(genres)
    })

    for genre in genres:
        genre_to_ids[genre].append(movie_id)

# Initialize Chroma
print("Initializing Chroma DB...")
client = chromadb.PersistentClient("./chroma")

try:
    client.delete_collection("movies_database")
except Exception:
    pass

collection = client.create_collection(name="movies_database")

# Upload in batches
batch_size = 500

for i in tqdm(range(0, len(documents), batch_size), desc="Uploading to Chroma"):
    collection.add(
        documents=documents[i:i+batch_size],
        metadatas=metadatas[i:i+batch_size],
        ids=ids[i:i+batch_size],
    )

print(f"\nAdded {collection.count()} movies to the 'movies' collection")
print(f"Indexed {len(genre_to_ids)} genres")
print("Example genres:", list(genre_to_ids.keys())[:10])
```

```python
results = collection.query(
    query_texts=["A romatic movie with my girlfriend"],
    n_results=3,
)

for i, doc in enumerate(results["documents"][0]):
    print(f"Result {i + 1}:")
    print(doc)
    print()
```
